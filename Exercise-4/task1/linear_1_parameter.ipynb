```python
from linear_plotter import *
from utils import *
%load_ext autoreload
%autoreload 2
alfa = 1
def getA(alfa):
    return np.asarray([[alfa,alfa],[-1/4,0]])
def f(time,pos,alfa):
    return getA(alfa)@pos
def f1(time,pos):
    return f(time,pos,alfa)
solution = scipy.integrate.solve_ivp(f1,(0,1),np.array([0,2]),dense_output = True)
w = 3
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
ax0 = plot_phase_portrait(getA(alfa), X, Y)
ax0.plot(solution.y[0], solution.y[1], c='red', label='Example trajectory')
ax0.legend()
ax0.set_aspect(1)
```
