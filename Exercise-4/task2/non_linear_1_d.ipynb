```python
import matplotlib.pyplot as plt

from one_d_plotter import *
from utils import *
%load_ext autoreload
%autoreload 2
def f1(x, alpha):
    return alpha - x*x

def f2(x, alpha):
    return alpha - 2*x*x - 3

plotter = OneDPlotter()

def plotalpha1(alpha):
    return plotter.plotalpha(alpha,f1,"system alpha - x*x")

def plotalpha2(alpha):
    return plotter.plotalpha(alpha,f2,"system alpha - 2*x*x - 3")

plotalpha1(1)
plotalpha2(1)
plotalpha1(-1)
plotalpha2(-1)
```

```python
alphas = np.arange(0,1000)/100
e1x = np.asarray(list(-np.sqrt(alphas[x]) for x in range(alphas.shape[0])))
e2x = np.asarray(list(np.sqrt(alphas[x]) for x in range(alphas.shape[0])))
def flatten(l):
    r = []
    for x in l:
        for y in x:
            r.append(y)
    return r
xs = np.arange(-1000,1000)/100
vs = np.asarray(flatten([[[alphas[alpha],xs[x]] for alpha in range(alphas.shape[0]) if np.abs(f1(xs[x],alphas[alpha])) < 0.1] for x in range(xs.shape[0])]))
plt.scatter(vs[:,0], vs[:,1])
plt.ylim(-10,10)
plt.xlim(-10,10)
plt.figure()
plt.plot(alphas, e1x)
plt.plot(alphas, e2x, '--')
plt.ylim(-10,10)
plt.xlim(-10,10)
```
