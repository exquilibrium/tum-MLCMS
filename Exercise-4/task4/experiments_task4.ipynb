Lorenz single trajectory

```python
# Set initial values and parameters
initial_values_original = np.array([10, 10, 10])
Tend = 1000
dt = 0.01
num_steps = int(Tend / dt)

# Simulate trajectories
trajectory_original = simulate_lorenz(initial_values_original, Tend, dt)

# Plot trajectories
viz.plot_trajectory_lorenz(trajectory_original)

#Plot trajectories in the x-y plane
viz.plot_trajectory_lorenz_xz(trajectory_original)
```

Lorenz two trajectories

```python
# Set initial values and parameters
initial_values_original = np.array([10, 10, 10])
initial_values_perturbed = np.array([10 + 1e-8, 10, 10])
Tend = 1000
dt = 0.01
num_steps = int(Tend / dt)

# Simulate trajectories
trajectory_original = simulate_lorenz(initial_values_original, Tend, dt)
trajectory_perturbed = simulate_lorenz(initial_values_perturbed, Tend, dt)

# Calculate the Euclidean distance between the two trajectories at each time step
distance = np.linalg.norm(trajectory_original - trajectory_perturbed, axis=1)

# Plot trajectories
viz.plot_multrajectories_lorenz(trajectory_original, trajectory_perturbed)


# Plot trajectories in the x-z plane
viz.plot_multrajectories_lorenz_xz(trajectory_original, trajectory_perturbed)

# Plot the Euclidean distance between the two trajectories over time
viz.plot_distance_lorenz(num_steps, dt, distance)
```
