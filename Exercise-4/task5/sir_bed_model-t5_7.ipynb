```python
from math import sqrt
from sir_model_visualization import *
```

```python
# Definition of set C^{+-}_{Delta} for Pitchfork Bifurcation
def f_plus(beta, mu0, mu1, d, nu):
    d0 = d+nu+mu0
    d1 = d+nu+mu1
    b = beta*(mu1-mu0)+d0*(d1-beta)
    b = b + sqrt(beta*d1*(mu1-mu0)*(d1-beta))
    b = b / ((beta-nu)*(d1*d1))
    return b
    
def f_minus(beta, mu0, mu1, d, nu):
    d0 = d+nu+mu0
    d1 = d+nu+mu1
    b = beta*(mu1-mu0)+d0*(d1-beta)
    b = b - sqrt(beta*d1*(mu1-mu0)*(d1-beta))
    b = b / ((beta-nu)*(d1*d1))
    return b
```

```python
A = 20  # "recruitment rate" (or birth rate) of susceptible population
d = 0.1  # per capita natural death rate
nu = 1  # per capita disease-induced death rate
mu0 = 10  # minimum recovery rate
mu1 = 10.45  # maximum recovery rate
beta = 11.5  # average number of adequate contacts per unit time with infectious individuals
#b = 0.022  # 0.01 # try to set this to 0.01, 0.020, ..., 0.022, ..., 0.03
```

```python
# Task 5.7 
# Pitchfork Bifurcation - See Theorem 4.1 (1)

# R_0 -> mu_1 is bifurcation parameter
# Set R_0 = 1
mu1 = beta - d - nu + 0
# Set b according to Theorem 4.1 (1)
b = A*(mu1-mu0)/(beta*(beta-nu))
sir_params = (A, d, nu, mu0, mu1, beta, b)

# Check parameters
print(f'Bifurcation possible {(beta>d+nu+mu0)}')
print(f'Reproduction number {R0(beta,d,nu,mu1)}')
print(f'b: {b} = {A*(mu1-mu0)/(beta*(beta-nu))}')
print(f'mu0: {mu0}, mu1: {mu1}')
print(f'f+ == f-: {f_plus(beta, mu0, mu1, d, nu)==f_minus(beta, mu0, mu1, d, nu)}')

# Create plots
SIM0 = (200,1e-9,0)
plotTrajectoryXY(init=SIM0, sirPar=sir_params, new_b=b, lines=True, opacity=0.1, ax0=0, ax1=1)
plotBifurcationT(init=SIM0, sirPar=sir_params, new_b=b, new_t=999)
plotTrajectory(init=SIM0, sirPar=sir_params, new_b=b)
plotBifurcation(init=SIM0, sirPar=sir_params, new_b=b)
```

```python
# Task 5.7 
# Forward Bifurcation - See Theorem 4.1 (1)

# R_0 -> mu_1 is bifurcation parameter
# Set R_0 = 1
mu1 = beta - d - nu + 0
# Set b according to  Fig.7 (a)
b = 0.07 #A*(mu1-mu0)/beta*(beta-nu)
sir_params = (A, d, nu, mu0, mu1, beta, b)

# Check parameters
print(f'Bifurcation possible {(beta>d+nu+mu0)}')
print(f'Reproduction number {R0(beta,d,nu,mu1)}')
print(f'b: {b} > {A*(mu1-mu0)/(beta*(beta-nu))}')
print(f'mu0: {mu0}, mu1: {mu1}')

# Create plots
SIM0 = (200,1e-9,0)
plotTrajectoryXY(init=SIM0, sirPar=sir_params, new_b=b, ax0=0, ax1=1)
plotBifurcationT(init=SIM0, sirPar=sir_params, new_b=b, new_t=999)
plotTrajectory(init=SIM0, sirPar=sir_params, new_b=b)
plotBifurcation(init=SIM0, sirPar=sir_params, new_b=b)
```
