Plot the equilibrium points for different ranges of alpha values, each associated with a different type of equilibrium point such as a focus or circular orbit.

```python
from utils import *
from zero_convergence_plotter import *
%load_ext autoreload
%autoreload 2
def f(x, alpha):
    return np.array([
        alpha * x[0] - x[1] - x[0] * (x[0] * x[0] + x[1] * x[1]),
        x[0] + alpha * x[1] - x[1] * (x[0] * x[0] + x[1] * x[1])])

pt = ZeroConvergencePlotter(f,3)
pt.plotalpha(-1, 0, "focus", True)
pt.plotalpha(0, 1, "focus", True)
pt.plotalpha(1, 2, "circular orbit", True)
```

Solve a 2D non-linear system for a given starting point and alpha value, and then plots the trajectory of the solution in a 2D space.

```python
import scipy


def plot_traj(sp, alpha):
    plt.figure(figsize=(10, 10))
    ax = plt.axes()
    solution = scipy.integrate.solve_ivp(lambda x, y: f(y, alpha), (0, 100), sp, dense_output=True)
    ax.plot(solution.y[0], solution.y[1], label=str(sp))
    ax.set_xlim((-2, 2))
    ax.set_ylim((-2, 2))


plot_traj(np.array([2, 0]), 1)
plot_traj(np.array([0.5, 0]), 1)
```
