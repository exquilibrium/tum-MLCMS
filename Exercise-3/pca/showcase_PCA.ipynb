```python
%load_ext autoreload
%autoreload 2
lines = list(str.strip(x) for x in open("pca_dataset.txt","r").readlines())
lines
```

```python
import numpy as np
data = np.asarray(list(tuple(float(y) for y in x.split(" ")) for x in lines if len(x) > 0))
data
```

```python
import model
import matplotlib.pyplot as plt
pca = model.ReversiblePCA()
pca.fit(data)
explained_var = pca.eigen_values / np.sum(pca.eigen_values)
print(tuple(f"{x[0]} : {float(x[1]):0.10f}" for x in enumerate(explained_var)))
```

```python
plt.scatter(data[:,0],data[:,1])
mean = np.mean(data,0)
for x in pca.principal_components:
    plt.arrow(mean[0], mean[1], x[0], x[1])
```

```python
pca.fit(data,False)
plt.scatter(data[:,0],data[:,1])
mean = np.mean(data,0)
for x in pca.principal_components:
    plt.arrow(mean[0], mean[1], x[0], x[1])
```
