```python
file_path = './Bottleneck_Data/uo-180-070.txt'

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the data file
column_names = ['ID', 'FRAME', 'X', 'Y', 'Z']
data = pd.read_csv(file_path, delimiter=' ', names=column_names)

# Create an empty list to store np.array for each ID
id_arrays = [np.array([]) for _ in range(data['ID'].max() + 1)]

# Group the data by ID and place the X, Y, Z columns of each group into the corresponding list item as np.array
grouped_data = data.groupby('ID')
for group_id, group_data in grouped_data:
    id_arrays[group_id] = group_data[['X', 'Y', 'Z']].to_numpy()

min_x = 10000
max_x = -10000
min_y = 10000
max_y = -10000


# Calculate range of the graph
for idx in id_arrays:
    for step in idx:
        if step[0]> max_x:
            max_x=step[0]
        if step[0]< min_x:
            min_x=step[0]
        if step[1]> max_y:
            max_y=step[1]
        if step[1]< min_y:
            min_y=step[1]
            
print(min_x, max_x)
print(min_y, max_y)


# Read the data into a pandas DataFrame
df = pd.read_csv(file_path, sep=' ', names=['ID', 'FRAME', 'X', 'Y', 'Z'])

# Get unique pedestrian IDs
pedestrian_ids = df['ID'].unique()

# Create a dictionary to map each pedestrian ID to a unique color
id_to_color = {pid: plt.cm.tab20b(i % 20) for i, pid in enumerate(pedestrian_ids)}

# Get unique frame numbers
frame_numbers = df['FRAME'].unique()

# Create a scatter plot for each frame
for frame in frame_numbers:
    frame_data = df[df['FRAME'] == frame]
    
    # Create a scatter plot for each pedestrian ID with consistent colors
    for pedestrian_id, group in frame_data.groupby('ID'):
        color = id_to_color[pedestrian_id]
        plt.scatter(group['X'], group['Y'], color=color, marker='o', s=90)

    # Set labels and title
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.title(f'Pedestrian Movement in 2D - Frame {frame}')
    
    plt.xlim(min_x, max_x)  # Replace min_x and max_x with your desired limits
    plt.ylim(min_y, max_y)  # Replace min_y and max_y with your desired limits

    # Save or show the plot for each frame
    plt.savefig(f'./img/frame_{frame}.png')  # You can save each frame as an image
    plt.close()  # Close the plot to start a new one for the next frame
```
